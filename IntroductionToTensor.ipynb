# **Tensor**

A tensor is a mathematical object used to store data. It is a generalization of scalars, vectors, and matrices to higher dimensions.

Think of a tensor as a **multi-dimensional** array.

Screenshot 2026-07-05 192236.png


**Why are tensors used**

Machine learning models work with large amounts of data, such as:

* Images
* Videos
* Audio
* Text

All of this data is represented as tensors so that mathematical operations can be performed efficiently.

**For example:**

* A grayscale image → 2D tensor (Height × Width)
* A color image → 3D tensor (Height × Width × RGB channels)
* A batch of images → 4D tensor
